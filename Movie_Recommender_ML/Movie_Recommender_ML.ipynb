```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import io
from google.colab import files
uploaded = files.upload()
```

```python
#load the dataset
data = pd.read_csv('movies.csv')
data.head()
```

```python
#importent features 
selection_feature =['genres','keywords','tagline','cast','director']
print(selection_feature)
```

```python
#replacing all the null values with null strings
for feature in selection_feature:
  data[feature] = data[feature].fillna(' ')#
```

```python
data[feature].head()
```

```python
#combining all the 5 features from the dataset
combine_feat = data['genres'] + data['keywords'] + data['tagline'] + data['cast'] + data['director']
print(combine_feat)
```

```python
#If you need the term frequency (term count) vectors for different tasks
#Convert a collection of raw documents to a matrix of TF-IDF features
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combine_feat)
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
movie_title = input('enter your fav. movie name:')
```

```python
list_of_all_titles = data['title'].tolist()
```

```python
find_closr_match = difflib.get_close_matches(movie_title,list_of_all_titles)
print(find_closr_match)
```

```python
close_match = find_closr_match[0]
```

```python
index_of_movie = data[data.title == close_match]['index'].values[0]
print(index_of_movie) #printing the index of the close matched movie
```

```python
#getting the list of similar movies with percent in list

similarity_score = list(enumerate(similarity[index_of_movie]))#
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1],reverse=True)
```

```python
sorted_similar_movies
```

```python
#print the names of the similar movies based on the index
print('\nUsers input : ',movie_title)

print('\nMovies suggested for you : ')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = data[data.index==index]['title'].values[0]
    if(i<10):
        print(i,'.',title_from_index)
        i+=1
        
```
